```python
def pre_process(data):
    df = data.lower()
    return df
```

```python
import string as str
def remove_specichar(data):
    symbols = {'~', ':', "'", '+', '[', '\\', '@', '^', '{', '%', '(', '-', '"', '*', '|', ',', '&', '<', '`', '}', '.', '_', '=', ']', '!', '>', ';', '?', '#', '$', ')', '/'}
    for i in data:
        print(i)
        if i in symbols:
            df = data.replace(i,'')

            return df
```

```python
import re
def sep_rem(data):
    df = re.sub(r"[^a-zA-Z0-9]"," ",data)
    df1 = re.sub(r"[-()\"#/@;:<>{}/''`+=~|.!?,]", " ", df)
    return df1
```
